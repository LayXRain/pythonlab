## 排序算法

### 冒泡排序

```python
list = [3,11,343,6,6,2,3,5,89,9,3,0,62]
#冒泡排序,将每个元素逐个放入有序区
def bubble_sort(li):
    for i in range(len(li)-1):
       for j in range(len(li)-i-1):
            if li[j]>li[j+1]:
               li[j],li[j+1]=li[j+1],li[j]
    return li

list1=bubble_sort(list)
print(list1)
```

### 选择排序

```python
#选择排序
def select_sort(li):
    for i in range(len(li) - 1):
        min_index = i
        for j in range(i + 1, len(li)):
            if li[j] < li[min_index]:
                min_index = j
        if min_index != i:
            li[i], li[min_index] = li[min_index], li[i]
nums = [1,23,53,412,32,45,11,24,242,66,88]
select_sort(nums)
print(nums)
```

### 插入排序

```python
#插入排序
def insert_sort(li):
    for i in range(1,len(li)):
        temp=li[i]
        j=i-1
        while j>=0  and li[j]>temp:
            li[j+1]=li[j]
            j-=1#从有序区从右往左看，依次#比较
            li[j+1]=temp

```

### 快速排序

```python

def classify(li,left,right):
        temp=li[left]#固定第一个值
        while left<right:
            while li[right]>=temp and left<right:#从右边往左找left<right控制循环终止
                right-=1
            li[left] = li[right]#交换左右值
            print(li)
            while li[left]<=temp and left <right:
                left+=1
            li[right]=li[left]
            print(li)
        li[left]=temp
        return left#temp值走到的位置

def quick_sort(li,left,right):
    if left<right:
        mid=classify(li,left,right)
        quick_sort(li,left,mid-1)
        quick_sort(li,mid+1,right)

print(list)
quick_sort(list,0,len(list)-1)
print(list)
```

### 堆排序

```python

def sift(li,low,high):
    i=low#根节点
    j=2*i+1#左孩子
    temp=li[low]
    while j<=high:
        if j+i<=high and li[j+1]>li[j]:#右孩子大于左孩子
            j=j+1 #将右孩子交换
        if li[j]>temp:#子节点如果一直比tmp大就一直向下找1
            li[i]=li[j]
            i=j
            j=2*i+1#子节点继续向下扩展
        else:#如果temp更大则将temp放进子节点
            li[i]=temp
            break
    else:#最底层子节点等于temp
        li[i]=temp

def heap_sort(li):
    n=len(i)
    for i in range((n-2)//2,-1,-1):
        sift(li,i,n-1)
        #建堆
    for i in range(n-1,-1,-1):
        #i指的是当前堆的最后一个元素
        li[0],li[i]=li[i],li[0]
        sift(li,0,i-1)

```

### 归并排序

```python
def merge(li, low, high, mid):
    i = low  # 左区间的起始索引
    j = mid + 1  # 右区间的起始索引
    ltmp = []  # 临时列表，存储合并后的结果
    while i <= mid and j <= high:
        if li[i] < li[j]:
            ltmp.append(li[i])
            i += 1
        else:
            ltmp.append(li[j])
            j += 1
    while i <= mid:
        ltmp.append(li[i])
        i += 1
    while j <= high:
        ltmp.append(li[j])
        j += 1
    li[low:high+1] = ltmp

def merge_sort(li, low, high):
    # 递归终止条件：当low >= high时，区间只有一个元素，无需排序
    if low < high:
        mid = (low + high) // 2  # 修正：mid计算放在if内部，逻辑更合理
        merge_sort(li, low, mid)  # 递归排序左区间[low, mid]
        merge_sort(li, mid+1, high)  # 递归排序右区间[mid+1, high]
        merge(li, low, high, mid)  # 合并两个有序区间


li = [1, 23, 4, 2, 4, 5, 1, 6]
merge_sort(li, 0, len(li)-1)
print(li)
```

### 希尔排序

```python

#希尔排序是基于将间隔为gap的列表排序，其中内涵插入排序
def insert_sort(li,gap):
    for i in range(gap,len(li)):
        temp=li[i]
        j=i-gap
        while j>=0  and li[j]>temp:
            li[j+gap]=li[j]
            j-=gap#从有序区从右往左看，依次#比较
            li[j+gap]=temp
def shell_sort(li):
    d=len(li)//2
    while d>=1:
        insert_sort(li,d)
        d//=2

list=[5,3,7,8,1,10,2,9,4]
shell_sort(list)
print(list)

```
